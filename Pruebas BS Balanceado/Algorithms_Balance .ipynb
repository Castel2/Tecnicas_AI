```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from pandas import set_option
import sklearn as sk
from sklearn.metrics import confusion_matrix

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report

from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import decomposition
from sklearn import preprocessing
```

```python
filename = '/content/drive/MyDrive/Proyecto Inteligencia/Resources/Thoracic Surgery Data - Balanceada.csv'
data_set = pd.read_csv(filename)
data_set
```

```python
data_set=sk.utils.shuffle(data_set)
data_set
```

```python
labels= data_set['Risk1Yr'].values
features = data_set.drop(['Risk1Yr'], axis=1)
```

```python
labels = labels.astype('int')
labels
```

```python
pca = decomposition.PCA(n_components=6)
pca.fit(features)
feature_PCA_Scaler = pca.transform(features)
```

```python
scaler = preprocessing.StandardScaler().fit(feature_PCA_Scaler)
feature_PCA_Scaler = scaler.transform(feature_PCA_Scaler)
```

```python
validation_size = 0.2
seed = 7
X_trainScaler, X_validationScaler, Y_trainScaler, Y_validationScaler = train_test_split(feature_PCA_Scaler, labels, 
                                                                                        test_size=validation_size, 
                                                                                        random_state=seed)
```

```python
KNN = KNeighborsClassifier()
KNN.fit(X_trainScaler, Y_trainScaler)
predictions = KNN.predict(X_validationScaler)
```

```python
print('precision en validacion')
print(accuracy_score(Y_validationScaler, predictions))
print('Matriz de confusion')
confusion_matrix(Y_validationScaler, predictions)
```

```python
LR = LogisticRegression()
LR.fit(X_trainScaler, Y_trainScaler)
predictions = LR.predict(X_validationScaler)
```

```python
print('precision en validacion')
print(accuracy_score(Y_validationScaler, predictions))
print('Matriz de confusion')
confusion_matrix(Y_validationScaler, predictions)
```

```python
DT = DecisionTreeClassifier()
DT.fit(X_trainScaler, Y_trainScaler)
predictions =DT.predict(X_validationScaler)
```

```python
print('precision en validacion')
print(accuracy_score(Y_validationScaler, predictions))
print('Matriz de confusion')
confusion_matrix(Y_validationScaler, predictions)
```

```python
SVM = SVC()
SVM.fit(X_trainScaler, Y_trainScaler)
predictions = SVM.predict(X_validationScaler)
```

```python
print('precision en validacion')
print(accuracy_score(Y_validationScaler, predictions))
print('Matriz de confusion')
confusion_matrix(Y_validationScaler, predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()
RF.fit(X_trainScaler, Y_trainScaler)
predictions = RF.predict(X_validationScaler)
```

```python
print('precision en validacion')
print(accuracy_score(Y_validationScaler, predictions))
print('Matriz de confusion')
confusion_matrix(Y_validationScaler, predictions)
```
