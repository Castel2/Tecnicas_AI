```python
import pandas as pd
import numpy
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
import seaborn as sns


filename = '/content/drive/MyDrive/Proyecto Inteligencia/Resources/Thoracic Surgery Data.csv'
data_set = pd.read_csv(filename)
data_set
```

```python
labels= data_set['Risk1Yr'].values
features = data_set.drop(['Risk1Yr'], axis=1)
```

```python
labels
```

```python
features
```

```python
labels.shape, features.shape, data_set.shape
```

```python
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(features, labels, test_size=validation_size, random_state=seed)
```

```python
X_train.shape, X_validation.shape, Y_train.shape, Y_validation.shape
```

```python
KNN=KNeighborsClassifier()
LR=LogisticRegression()
```

```python
KNN.fit(X_train,Y_train)
```

```python
KNN.score(X_validation,Y_validation)
```

```python
predic=KNN.predict(X_validation)
predic
```

```python
predic==Y_validation
```

```python
print(accuracy_score(Y_validation, predic))
print(confusion_matrix(Y_validation, predic))
print(classification_report(Y_validation, predic))
```

```python
kfold = model_selection.KFold(n_splits=10, random_state=seed)
cv_results = model_selection.cross_val_score(KNN, features, labels, cv=kfold, scoring='accuracy')
msg = "%s: %f (%f)" % ('KNN', cv_results.mean(), cv_results.std())
print(msg)
```
